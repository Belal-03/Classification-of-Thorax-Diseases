# **• Data Augmentation Techniques**:
## <center>**○ Random Zoom in or Zoom out with ± 0.001%**
## <center>**○ Random change the brightness by ± 0.01%**
## <center>**○ Random change the contrast by ± 0.02%**

```python
data_augmentation_net = Sequential([
    RandomZoom(height_factor=(-0.001, 0.001), width_factor=(-0.001, 0.001)),
    RandomBrightness(factor=0.01),
    RandomContrast(0.02)
])
```

# **• Convert the output from (data_augmentation_net) to a NumPy array of integer values.**

```python
def convert_to_array(tensor):
    tensor = [np.array(tensor[i]) for i in range(len(tensor))]    # convert it to numpy array
    tensor = [cv.convertScaleAbs(tensor[i]) for i in range(len(tensor))] # convert the values in array to int and absolute
    return np.array(tensor)
```

```python
def do_aug(images ,training = True):
  aug_images = data_augmentation_net(images ,training = training)
  aug_images = convert_to_array(aug_images)
  return aug_images
```

```python
def make_dims_for_image(img):
    grayscale_image = np.expand_dims(img, axis=-1)
    rgb_image = np.repeat(grayscale_image, 3, axis=-1)
    return rgb_image.astype(np.float16)
```

```python
df_all["processed_arrays"] = df_all["Image_array"].apply(lambda x : make_dims_for_image(x))
```
